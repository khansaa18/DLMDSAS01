```python
import matplotlib.pyplot as plt
import numpy as np

# Given parameters
ξ1 = 0  # Indicates a Bernoulli distribution
ξ2 = 0.45  # Probability of "for" in a single Bernoulli trial

# Values for visualization
categories = ['Against', 'For']
probabilities = [1 - ξ2, ξ2]
percentage_for = ξ2 * 100

# Visualization
plt.figure(figsize=(8, 6))
plt.bar(categories, probabilities, color=['red', 'green'])
plt.xlabel('Vote Outcome')
plt.ylabel('Probability')
plt.title('Bernoulli Distribution for Voting Outcomes')
plt.text(0.5, 0.5, f'{percentage_for:.2f}%', color='white', fontsize=12,
         ha='center', va='center')
plt.ylim(0, 1)
plt.show()
```
