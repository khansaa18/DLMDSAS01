```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma

# Given values
xi17 = 77
xi18 = 79
xi19 = 52.06
sample_size = 10
alpha = 3

# Generate a random sample from a gamma distribution
np.random.seed(42)
sample_data = np.random.gamma(shape=alpha, scale=1/xi19, size=sample_size)

# Posterior distribution parameters
posterior_alpha = alpha + sample_size * xi17
posterior_beta = 1 / (1/xi18 + np.sum(sample_data))

# Posterior distribution
posterior_distribution = gamma(a=posterior_alpha, scale=1/posterior_beta)

# Plot the prior and posterior distributions
theta_values = np.linspace(0, 100, 1000)
prior_values = gamma.pdf(theta_values, a=xi17, scale=1/xi18)
posterior_values = posterior_distribution.pdf(theta_values)

plt.plot(theta_values, prior_values, label='Prior Distribution', linestyle='--')
plt.plot(theta_values, posterior_values, label='Posterior Distribution')
plt.title('Prior and Posterior Distributions of $\\theta$')
plt.xlabel('$\\theta$')
plt.ylabel('Probability Density')
plt.legend()
plt.show()

# Bayes point estimates
mean_estimate = posterior_distribution.mean()
mode_estimate = theta_values[np.argmax(posterior_values)]

print(f'Bayes Point Estimate (Mean): {mean_estimate:.4f}')
print(f'Bayes Point Estimate (Mode): {mode_estimate:.4f}')
```
